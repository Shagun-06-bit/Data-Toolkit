1. What is NumPy, and why is it widely used in Python?

2. How does broadcasting work in NumPy?
- NumPy in Python is a powerful feature that allows arrays of different shapes and sizes to be used together in mathematical operations without explicitely reshaping them.
Some Broadcasting rules are:
- NumPy compares the shapes of the arrays from right to left to make broadcasting work:
- If the dimensions of the two arrays are equal then no error will occur.
- If one of the arrays are 1 then that is not a problem.
- But if the dimensions of the two arrays are unequal and neither of them are 1 then an error will occur.

```python
#Broadcasting example
import numpy as np
a=np.array([4,5,6])
b=np.array(5) # b is treated as [5,5,5]
print(a+b)
```

4. Explain the use of the groupby() method in Pandas.
- groupby() function is used in Pandas to split the data into groups and then perform sum mathematical operations like sum(), mean() and then combine the results. It is very useful in analyzing and summarising data based on one or more categorical column.

```python
import pandas as pd
data={'Department': ['IT', "HR",'Finance','IT','Sales','HR'],
      'Salary':[20000, 30000, 45000, 80000,35000, 26000],
      'Employee':['A','B','C','D','E','F']}
d=pd.DataFrame(data)
group=d.groupby('Department')['Salary'].sum()
group
```

21. A Why is NumPy’s array slicing faster than Python’s list slicing?

Some of the reasons why NumPy arrays are faster are:


1. Homogenous data: NumPy arrays store homogenous data which helps in faster computation and occupy less memory space.
2. Contiguous Memory Layout: Arrays are stored in contiguous memory locations thus making array indexing and slicing faster.
3. No type checking overhead: Lists check data types at run time whereas NumPy arrays don't they already know it.

```python
'''
4. A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
'''


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data={'Maths':[89,67,80,94,50],
      'Physics':[26,78,95,67,80],
      'Chemistry':[69,45,82,78,99]}
df=pd.DataFrame(data)
corr=df.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
